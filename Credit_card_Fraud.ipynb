```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the credit card dataset to the pandas dataframe
data = pd.read_csv('/content/creditcard.csv')
```

```python
#printing top 5 rows of the dataset using default head() command
data.head()
```

```python
#printing bottom 5 rows of the dataset using default tail() command
data.tail()
```

```python
#dataset information
data.info()
```

```python
data.isnull().sum()
```

```python
#dealing with the missing values in the dataset
data = data.interpolate(method ='linear' , limit_direction = 'forward')
data
```

```python
data.isnull().sum()
```

```python
#distribution of legit transaction and fraudulent transaction
data['Class'].value_counts()
```

HERE:-                          
LABEL 0 REPRESENTS VALID TRANSACTIONS                                
LABEL 1 REPRESENTS FRAUD TRANSACTION


```python
# seperating the data for analysis
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
fraud
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comparing values of both transaction
data.groupby('Class').mean()
```

Under-Sampling

building sample dataset containing similar distribution of legit and fraud transaction

```python
legit_sample = legit.sample(n = 492)
```

Concatenating the 2 dataframe

```python
new_ds = pd.concat([legit_sample,fraud], axis = 0)
```

```python
new_ds.head()
```

```python
new_ds.tail()
```

```python
new_ds['Class'].value_counts()  #here we have a sample pair for both fraud and legit transactions
```

```python
new_ds.groupby('Class').mean() #here the mean of legit transactions is almost similar to that of entire mean of legit transactions
```

Splitting data into features and targets

```python
X = new_ds.drop(columns = 'Class', axis = 1)
Y = new_ds['Class']
```

```python
print(X)
```

```python
print(Y)
```

splitting data into training and testing data

```python
X_train , X_test , Y_train, Y_test = train_test_split(X,Y,test_size = 0.2 , stratify = Y , random_state= 3)
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

Model Training using Logistic Regression

```python
model = LogisticRegression()
#training model with training data
model.fit(X_train , Y_train)
```

Model Evaluation based on accuracy score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("accuracy on training data: ", training_data_accuracy)
```

Here Accuracy score is 91.86% which means out of 100 predictions almost 92 predictions are accurate

```python
#accuracy on test data (unseen data)
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction , Y_test)
print('Accuracy on test data: ', test_data_accuracy)
```

Also above 90.8%
